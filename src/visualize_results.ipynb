```python
import os
import matplotlib.pyplot as plt
import requests
import json

CERT_PATH = os.path.join('../data', 'cert.pem')    

# Make a request to your server to get the metrics
response = requests.post('https://localhost:443/feed', json={}, verify=CERT_PATH)
metrics = json.loads(response.headers['X-Response-Data'])['metrics']

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(metrics['train_losses'], label='Training Loss')
plt.plot(metrics['val_losses'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Plot training and validation Mean Absolute Error
plt.figure(figsize=(12, 6))
plt.plot(metrics['train_maes'], label='Training MAE')
plt.plot(metrics['val_maes'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('Mean Absolute Error')
plt.title('Training and Validation Mean Absolute Error')
plt.legend()
plt.show()

# Plot training and validation Root Mean Squared Error
plt.figure(figsize=(12, 6))
plt.plot(metrics['train_rmses'], label='Training RMSE')
plt.plot(metrics['val_rmses'], label='Validation RMSE')
plt.xlabel('Epoch')
plt.ylabel('Root Mean Squared Error')
plt.title('Training and Validation Root Mean Squared Error')
plt.legend()
plt.show()
```
